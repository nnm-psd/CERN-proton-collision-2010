```python
#Reading data
import numpy as np
import pandas as pd
data= pd.read_csv('../input/multijet-primary-dataset/MultiJetRun2010B.csv')
data.head()
```

```python
#No null or categorical data
data.info()
```

```python
data.describe()
```

```python
#importing libraries
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
```

```python
#seaborn to plot between any two columns, nJet which is the target is highlighted
sns.pairplot(
    data,hue="nJets",
    x_vars=["MR", "HT", "E1","E2","Rsq","MET"],
    y_vars=["MR", "HT", "E1","E2","Rsq","MET",],
)
```

```python
pd.set_option('precision',2)
plt.figure(figsize=(25, 15))
sns.heatmap(data.drop(['Run','Event'],axis=1).corr(), annot=True) #Run and Event are not variable compatible for machine learning
plt.suptitle("Pearson Correlation Heatmap")
#plt.show()
```

```python
#split data, target is nJets
#save run column for the final table
#feature engineering

y = data['nJets']
X = data.drop(columns=['nJets','Event','Lumi','nBJets','Pz2','MET','Pz1'])

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
run = X_test['Run'].copy()
X_test = X_test.drop('Run',axis=1)
X_train = X_train.drop('Run',axis=1)
```

```python
print(len(X))
print(len(X_train)+len(X_test))
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.25, random_state=101)
```

```python
print(len(X_train))
print(len(X_test))
print(len(X_valid))
print(len(X))
print(len(X_train)+len(X_test)+len(X_valid))
```

VERSION RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
rf_r = RandomForestRegressor(max_depth=50, n_estimators=100, random_state=1)
rf_r.fit(X_train,y_train)
```

```python
X.head()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.model_selection import cross_val_score
cross_val_scores = cross_val_score(RandomForestRegressor(max_depth=1, n_estimators=100, random_state=1),\
                                   X, y, scoring='neg_mean_squared_error', cv=5)
```

```python
# cheking training accuracy
y_pred = rf_r.predict(X_train)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_train,predictions)
accuracy
```

```python
# cheking initial test accuracy
y_pred = rf_r.predict(X_test)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test,predictions)
accuracy
```

```python
feature_importance = np.array(rf_r.feature_importances_)
feature_names = np.array(X_train.columns)
data1={'feature_names':feature_names,'feature_importance':feature_importance}
data_plt = pd.DataFrame(data1)
sns.barplot(x=data_plt['feature_importance'], y=data_plt['feature_names'])
plt.xlabel('feature_importance')
plt.ylabel('feature_names')
plt.show()
```

```python
table = pd.DataFrame({'Run':run,'nJets':predictions})
table
```

```python
table.to_csv('CERN proton collision solution RForest.csv',index=False)
```
